```python
def cap(s):
    words = s.split()
    result = []

    for w in words:
        first = w[0].upper()
        rest = w[1:].lower()
        result.append(first + rest)

    return " ".join(result)

print(cap("hello python numpy"))
```

```python
# Q3. Create a function that checks if a number is prime (optimize with √n logic).

def prime(n):
    if n < 2:
        return "Not Prime"
    
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Not Prime"
    
    return "Prime"

prime(2)
```

```python
# Q6. Convert a nested list into a flat list using:

#loop

f = [1, [2, 3], [4, [5, 6]], 7]

flat = []
stack = f[:]            # copy list

while stack:
    item = stack.pop(0) # take first element
    if isinstance(item, list):
        stack = item + stack   # expand the list into the front
    else:
        flat.append(item)

print(flat)

```

```python
# List Comprehension

def flatten_comp(lst):
    return [
        y
        for item in lst
        for y in (
            flatten_comp(item)      # recursive flatten
            if isinstance(item, list)
            else [item]             # wrap non-list in list
        )
    ]

# example
data = [1, [2, 3], [4, [5, 6]], 7]
print(flatten_comp(data))
```

```python
# Recursion

def flatten_rec(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten_rec(item))
        else:
            result.append(item)
    return result

print(flatten_rec([1, [2, 3], [4, [5, 6]], 7]))
```

```python
# Generator

def flatten_gen(lst):
    for item in lst:
        if isinstance(item, list):
            yield from flatten_gen(item)
        else:
            yield item

print(list(flatten_gen([1, [2, 3], [4, [5, 6]], 7])))
```

```python
# Q10. Given a list of names, group names starting with same letter into a dict.

def group_names(names):
    groups = {}

    for name in names:
        key = name[0].upper()
        if key not in groups:
            groups[key] = []
        groups[key].append(name)
    
    return groups

names = ["Aman", "Ajay", "Ankit", "Bunty", "Bharat", "Chirag", "Charu", "Deepak"]
print(group_names(names))
```

```python
# Q11. Implement a calculator class (add/sub/mul/div) with method overloading style.

class Calculator:

    def add(self, *args):
        return sum(args)

    def sub(self, *args):
        if len(args) == 1:
            return args[0]
        result = args[0]
        for x in args[1:]:
            result -= x
        return result

    def mul(self, *args):
        result = 1
        for x in args:
            result *= x
        return result

    def div(self, *args):
        if len(args) < 2:
            return "Need at least 2 values"
        result = args[0]
        for x in args[1:]:
            result /= x
        return result

calc = Calculator()

print(calc.add(10, 20))
print(calc.add(1, 2, 3, 4))

print(calc.sub(10, 2))
print(calc.mul(2, 3, 4))

print(calc.div(100, 2, 5))
```

```python
# Q12. Create a function that accepts unknown arguments (*args, **kwargs) and prints patterns.

def pattern_printer(*args, **kwargs):
    print("ARGS (Unknown Positional Values):")
    for i, val in enumerate(args, start=1):
        print(f"{i}.{val}")

    print("\nKWARGS (Unknown Named Values):")
    for key,value in kwargs.items():
        print(f"{key} : {value}")

    print("\nPATTERN OUTPUT:")
    for val in args:
        if isinstance(val,int):
            print("*" * val)

    for key, value in kwargs.items():
        if isinstance(value,int):
            print(f"{key}:" + "#"*value )
            # print(f"{key}: " + "#" * value)

pattern_printer(3, 5, name="kunal", stars=7)
```

```python
# Q13. Demonstrate encapsulation using private variables (__var).

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposite(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient Funds")

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposite(5000)
acc.withdraw(2000)
print(acc.get_balance())
```

```python
# Q14. Write a decorator to time any function.

import time

def timer(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"Time taken: {end - start:.6f} sec")
        return result
    return wrapper


@timer
def test(n):
    for i in range(n):
        pass

test(1000000)
    
```

```python
# Q16. Read a CSV file without using pandas and compute average of a column.

import csv

def average_column(filename, column_name):
    total = 0
    count = 0

    with open(filename, "r") as file:
        reader = csv.DictReader(file)

        for row in reader:
            value = float(row[column_name])   # convert string → number
            total += value
            count += 1

    return total / count if count > 0 else 0


# Example usage
avg = average_column("marks.csv", "score")
print("Average Score:", avg)
```

```python
# Q18. Log errors to a file named "error.log".

import logging

# Configure the logging system
logging.basicConfig(
    filename="error.log",          # log file name
    level=logging.ERROR,           # only log errors+
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def do_divide(a, b):
    try:
        return a / b
    except Exception as e:
        logging.error(f"Error occurred: {e}")   # log the error
        print("Error logged in error.log")      # message to user
        return None

# Test
print(do_divide(10, 0))
print(do_divide(5, "abc"))
```

```python
def merge_files(file1, file2, output):
    try:
        with open(file1, "r") as f1, open(file2, "r") as f2:
            data1 = f1.read()
            data2 = f2.read()

        with open(output, "w") as out:
            out.write(data1 + "\n" + data2)

        print("Files merged successfully!")

    except FileNotFoundError:
        print("Error: One or both files not found.")
    except Exception as e:
        print("Unexpected Error:", e)


merge_files("file1.txt", "file2.txt", "output.txt")
```

```python
# Q21. Explain shallow vs deep copy using copy module.

import copy

a = [1, 2, [3, 4]]

b = copy.copy(a)

print(id(a), id(b))          # different
print(id(a[2]), id(b[2]))    # SAME → shallow copy
```

```python
# Q22. Show lambda + map + filter + reduce usage on a list of numbers.

from functools import reduce


nums = [1, 2, 3, 4, 5, 6]

sq = list(map(lambda x: x *x ,nums))
even = list(filter(lambda x: x%2 == 0, nums))
product = reduce(lambda a,b : a*b,nums)

print("Squares:", sq)
print("Evens:", even)
print("Product:", product)
```

```python
# Q24. Implement a custom iterator class.

class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.limit:
            value = self.current
            self.current += 1
            return value
        else:
            raise StopIteration


c = Counter(5)

for num in c:
    print(num)
```

```python
import time

class TimeLogger:
    def __init__(self, filename, mode="r"):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.start = time.time()     # start timer
        self.file = open(self.filename, self.mode)
        return self.file             # give file object to user

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.file.close()            # close file
        end = time.time()
        print(f"Execution Time: {end - self.start:.6f} sec")

        # Returning False → errors (if any) are not suppressed
        return False
```
