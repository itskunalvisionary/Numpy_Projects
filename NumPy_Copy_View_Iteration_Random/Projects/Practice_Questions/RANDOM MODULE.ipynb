```python
import numpy as np
import random
```

```python
# ❓ Q16 — Generate 10 random floats (0–1)

np.random.rand(10)
```

```python
# ❓ Q17 — Generate random integers 50–100 (size=20)

np.random.randint(50,101,20)
```

```python
# ❓ Q18 — Generate normal distribution (mean=0, std=1)

a = np.random.randn(10)
```

```python
# ❓ Q19 — Shuffle array IN-PLACE

arr = np.arange(10)
print(arr)
np.random.shuffle(arr)
print(arr)
print(arr)
```

```python
# ❓ Q20 — Create shuffled copy (original safe)

arr = np.arange(10)
print(arr)

new = np.random.permutation(arr)
print(new)

print(arr)
```

```python
# ❓ Q21 — Random choice from list

cities = np.array(["Delhi","Mumbai","Pune","Goa"])

np.random.choice(cities, size=5)
```

```python
# ❓ Q22 — Random sample WITHOUT replacement

np.random.choice(np.arange(16), size=3,replace=False)
```

```python
# ❓ Q23 — Set seed for reproducible results
# ❓ Q24 — Generate 2D random matrix (3×4)

np.random.seed(42)
np.random.randint(3,100,4)
```
