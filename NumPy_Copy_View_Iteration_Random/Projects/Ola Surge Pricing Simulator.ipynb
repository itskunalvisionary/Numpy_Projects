**Dataset Generation**

```python
# Base fare per ride:
base = np.random.randint(100, 300, 20)
base
```

```python
# Surge multipliers:
surge = np.random.uniform(1.0, 2.5, 20)
surge
```

**Apply broadcasting**

```python
final = base * surge
final
```

**Find expensive rides (>400)**

```python
expensive_ride = final[final>400]
expensive_ride
```

**ðŸŽ¯ STEP 5 â€” Print each ride (index + fare)**

```python
for index, fare in np.ndenumerate(final):
    print(index[0],round(fare,2))
```
