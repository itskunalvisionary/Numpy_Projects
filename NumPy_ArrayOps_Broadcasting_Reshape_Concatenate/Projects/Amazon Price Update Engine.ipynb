Tasks:

```python
prices = np.array([200,300,150,400,500,600])
discount = np.array([10])  # flat 10%
```

**1. Apply discount**

```python
final = prices * 0.90
print(final)
```

**2. Separate cheap (<300) and costly (>=300)**

```python
cheap = final[final < 300]
costly = final[final >= 300]


print("Cheap items:", cheap)
print("Costly items:", costly)
```

**3. Combine both into final list**

```python
combined = np.concatenate((cheap,costly))
combined
```

**4. Reshape into 2-column dashboard**

```python
dashboard = combined.reshape(-1,2)
dashboard
```
