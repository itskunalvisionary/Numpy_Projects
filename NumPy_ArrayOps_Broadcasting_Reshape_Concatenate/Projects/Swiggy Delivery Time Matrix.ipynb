Reshape into (4,2) matrix

Find morning vs evening average

Extract fast deliveries (<35)

Extract slow deliveries (>50)

```python
import numpy as np
times = np.array([30,25,40,50,35,28,45,55])
```

**1. Reshape into weekly matrix (4 days Ã— 2 slots)**

```python
matrix = times.reshape(4,2)
print(matrix)
```

**2. Morning vs Evening avg**

```python
morning = matrix[:,0].mean()  # 0 for morning column and : for all rows
evening = matrix[:,1].mean()  # 1 for evening column and : for all rows
print("Morning avg:", morning)
print("Evening avg:", evening)
```

**3. Fast deliveries (<35)**

```python
fast = times[times<=35]
print("Fast deliveries (<35 mins):", fast)
```

**4. Slow deliveries (>50)**

```python
slow = times[times>=50]
print("Slow deliveries (>50 mins):", slow)
```
