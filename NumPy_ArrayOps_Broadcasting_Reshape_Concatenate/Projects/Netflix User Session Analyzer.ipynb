Tasks:

Reshape into (4,3)

Extract binge sessions (>100)

Split dataset into first half & second half

Concatenate reversed sessions with original

```python
import numpy as np
```

```python
#Dataset

sessions = np.array([45,60,30,200,150,20,90,120,30,15,75,180])
```

**1. Reshape into 4 weeks Ã— 3 sessions**

```python
weeks = sessions.reshape(4,3)
print(weeks)
```

**2. Find binge (>100)**

```python
binge = sessions[sessions>=100]
print("Binge sessions (>100 mins):", binge)
```

**3. Split dataset into first half & second half**

```python
part1, part2 = np.split(sessions,2)
print("First half sessions:", part1)
print("Second half sessions:", part2)
```

**4. Concatenate reversed session + original**

```python
combined = np.concatenate((sessions[::-1],sessions))
print("Concatenated reversed and original sessions:", combined)
```
