```python
import numpy as np
```

```python
# ❓ Q17 — concatenate 1D

a = np.array([1,2,3])
b = np.array([4,5,6])
np.concatenate((a,b))
```

```python
# ❓ Q18 — vertical stack

x = np.array([1,2,3])
y = np.array([4,5,6])
np.vstack((x,y))  # vertical stack
```

```python
# ❓ Q19 — horizontal stack

x = np.array([1,2,3])
y = np.array([4,5,6])
np.hstack((x,y))  # horizontal stack
```

```python
# ❓ Q20 — split array

arr = np.array([10,20,30,40,50,60])
np.split(arr,3)
```

```python
# ❓ Q21 — hsplit

m = np.array([[1,2,3],[4,5,6]])
np.hsplit(m,3)  # split into 3 arrays horizontally

# ❓ Q22 — vsplit
np.vsplit(m,2)  # split into 2 arrays vertically
```
