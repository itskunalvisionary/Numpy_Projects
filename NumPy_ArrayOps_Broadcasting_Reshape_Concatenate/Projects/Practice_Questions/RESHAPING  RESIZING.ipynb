```python
import numpy as np
```

```python
# ❓ Q11 — reshape 1D → 2D

arr = np.arange(1,13)

arr.reshape(3,4)  # 3 rows, 4 columns
```

```python
# ❓ Q12 — reshape 1D → 3D

np.arange(1,9)
np.arange(1,9).reshape(2,2,2)
```

```python
# ❓ Q13 — flatten array

m = np.array([[1,2,3],[4,5,6]])
m.flatten()
```

```python
# ❓ Q14 — resize array

x = np.array([1,2,3])
np.resize(x, 6)
```

```python
# ❓ Q15 — transpose matrix

m = np.array([[1,2],[3,4],[5,6]])
m.T
```

```python
# ❓ Q16 — reshape using -1

arr = np.arange(12)
arr.reshape(2,-1)  # 2 rows, automatically calculate columns
```
