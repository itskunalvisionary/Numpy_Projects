```python
import numpy as np
```

```python
""" 
Task:

Add

Subtract

Multiply

Divide

"""


a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

# Addition
add = a +b
print("Addition:", add)
# Subtraction
sub = a - b 
print("Subtraction:", sub)
# Multiplication    
mul = a * b
print("Multiplication:", mul)
# Division
div = a / b
print("Division:", div)
```

```python
# ❓ Q2 — Scalar operations

arr = np.array([5, 10, 15, 20])

# Add 10 to each element
add_scalar = arr + 10
print("Add 10:", add_scalar)
# Subtract 5 from each element
sub_scalar = arr - 5    
print("Subtract 5:", sub_scalar)
# Multiply each element by 3
mul_scalar = arr * 3
print("Multiply by 3:", mul_scalar)
# Divide each element by 2 
div_scalar = arr / 2
print("Divide by 2:", div_scalar)
```

```python
# ❓ Q3 — Comparison Operations

"""   
Find:

scores >= 50

scores < 40

scores == 90

"""


marks = np.array([45, 67, 89, 23, 78, 90, 56])

# Greater than or equal to 50
great = marks >= 50
# great = marks[marks >= 50]
print("Scores >= 50:", great)
# Less than 40
less = marks < 40
print("Scores < 40:", less)
# Equal to 90
equal = marks == 90
print("Scores == 90:", equal)
```

```python
# ❓ Q4 — Logical AND / OR

"""   
Find numbers:

between 25 and 75

< 30 OR > 80

"""

arr = np.array([12, 25, 30, 50, 75, 90])

# Between 25 and 75
mid = arr[(arr >= 25) & (arr <= 75)]
print(mid)
side = arr[(arr < 30) | (arr >80)]
print(side)
```

```python
# ❓ Q5 — Aggregate Functions (sum, mean, min, max)

sales = np.array([200, 300, 250, 400, 350])

total_sales = np.sum(sales)
print("Total Sales:", total_sales)
average_sales = np.mean(sales)
print("Average Sales:", average_sales)
min_sales = np.min(sales)
print("Minimum Sales:", min_sales)
max_sales = np.max(sales)
print("Maximum Sales:", max_sales)
```

```python
# ❓ Q7 — Broadcasting Magic

prices = np.array([100, 200, 300])
tax = np.array([5])

final_prices = prices + tax
print("Final Prices with Tax:", final_prices)
```
