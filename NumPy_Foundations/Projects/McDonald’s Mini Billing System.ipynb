**McDonald's menu prices:**

```python
import numpy as np
prices = np.array([120,80,150,200,50])
# Items: Burger, Fries, Shake, Meal, Coke
order = [0,1,4,3]
```

**1. Ordered item prices**

```python
items = prices[order]
items

# prices[[0,1,4,3]] #Alternative
```

**2. Slice first 3 items**

```python
first_3 = prices[:3]
first_3
```

**3. Extract costly items from Order (>100)**

```python
costly = items[items>=100]
costly
```

**4. Extract costly items from Menu (>100)**

```python
costly = prices[prices>100]
costly
```

**5. Calculate bill**

```python
total_bill = np.sum(items)
print(total_bill)
```

**5. Discount check**

```python
if total_bill > 400:
    print("Discount Allow")

else:
    print("No Discount")
```
