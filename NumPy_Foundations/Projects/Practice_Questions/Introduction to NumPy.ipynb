```python
# Q1 — Temperature List → NumPy Array + Speed Difference
import numpy as np

# This is Python List of temperatures in Celsius
temps = [29.5, 30.1, 31.2, 28.9, 27.8, 29.9, 30.5]

# Convert the Python List to a NumPy Array
temp_array = np.array(temps)
print("NumPy Array of Temperatures:", temp_array)
print("-"*50)
avg = temp_array.mean()
print(avg)
print("-"*50)
Total = temp_array.sum()
print(Total)
```

```python
# Q5 — Convert car speeds km/h → m/s

speeds = [45, 55, 60, 52, 70, 68, 61, 74, 58, 49]

arr = np.array(speeds)
ms = arr/3.6
print("Speeds in m/s:", ms)

print("-"*50)

for speed in speeds:
    convert = speed / 3.6
    print(convert)
```
