```python
import numpy as np
```

```python
# Q17 — slices

data = np.array([10,20,30,40,50,60,70])
first3 = data[:3]
last2 = data[-2:]
middle = data[2:5]
print("First 3 elements:", first3)
print("Last 2 elements:", last2)
print("Middle elements (index 2 to 4):", middle)
```

```python
# Q18 — reverse

data = np.array([10,20,30,40,50])
reversed_data = data[::-1]
reversed_data
```

```python
# Q19 — 2D matrix slicing

m = np.array([[10,20,30],[40,50,60],[70,80,90]])

row1 = m[0]
col2 = m[:,1]
sub = m[1:3, 1:3]
print("First row:", row1)
print("Second column:", col2)
print("Sub-matrix (rows 1-2, cols 1-2):", sub)
```

```python
# Q20 — boolean mask

ages = np.array([15,22,17,45,33,28,12])

adults = ages[ages >= 18]
print(adults)
minor = ages[ages < 18]
print(minor)
mid = ages[(ages>=18) & (ages<=30)]
print(mid)
```

```python
# Q21 — fancy indexing

arr = np.array([100,200,300,400,500])
res = arr[[0,2,4]] # fancy indexing to get elements at index 0, 2, and 4
res
```

```python
# Q22 — modify using fancy indexing

arr = np.array([10,20,30,40,50])
arr[[1,3]] +=100
arr
```
